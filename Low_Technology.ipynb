```python
# Step 1: Group by Year and calculate the sum for the whole world (all regions combined)
yearly_world_sums = (
    low_technology_exports.groupby('Year')[['C- Actual Export of Low Technology', 'C- Mirrored Exports of Low Technology']]
    .sum()
    .reset_index()
)

# Step 2: Rename columns for clarity (optional)
yearly_world_sums.rename(columns={
    'C- Actual Export of Low Technology': 'W- Actual Export of Low Technology',
    'C- Mirrored Exports of Low Technology': 'W- Mirrored Exports of Low Technology'
}, inplace=True)

# Step 3: Merge the yearly sums back into the original DataFrame
low_technology_exports = pd.merge(
    low_technology_exports,
    yearly_world_sums,
    on='Year',  # Merge based on Year
    how='left'
)

# Step 4: Verify the result
print(low_technology_exports.head())

```

```python
# Calculate C-APE, R-APE, and W-APE using the provided formulae
low_technology_exports['C-APE'] = abs(((low_technology_exports['C- Actual Export of Low Technology'] - low_technology_exports['C- Mirrored Exports of Low Technology']) / low_technology_exports['C- Actual Export of Low Technology']) * 100)

low_technology_exports['R-APE'] = abs(((low_technology_exports['R- Actual Export of Low Technology'] - low_technology_exports['R- Mirrored Exports of Low Technology']) / low_technology_exports['R- Actual Export of Low Technology']) * 100)

low_technology_exports['W-APE'] = abs(((low_technology_exports['W- Actual Export of Low Technology'] - low_technology_exports['W- Mirrored Exports of Low Technology']) / low_technology_exports['W- Actual Export of Low Technology']) * 100)

# Verify the result
print(low_technology_exports.head())
```

```python
# Calculate C-MAPE (Grouped by Country)
def calculate_mape(x):
    valid_values = x.dropna()
    if valid_values.empty:  # Avoid division by zero if there are no valid values
        return 0
    return (valid_values.sum() / valid_values.count()) if valid_values.count() > 0 else 0

country_mape = low_technology_exports.groupby('Country')['C-APE'].apply(calculate_mape).reset_index()
country_mape.columns = ['Country', 'C-MAPE']

# Merge C-MAPE back to the original DataFrame
low_technology_exports = low_technology_exports.merge(country_mape, on='Country', how='left')

# Calculate R-MAPE (Grouped by Region)
region_mape = low_technology_exports.groupby('Region')['R-APE'].apply(calculate_mape).reset_index()
region_mape.columns = ['Region', 'R-MAPE']

# Merge R-MAPE back to the original DataFrame
low_technology_exports = low_technology_exports.merge(region_mape, on='Region', how='left')

# Calculate W-MAPE (For the whole world)
world_mape = low_technology_exports['W-APE'].dropna().sum() / low_technology_exports['W-APE'].dropna().count() if low_technology_exports['W-APE'].dropna().count() > 0 else 0

# Add W-MAPE to all rows
low_technology_exports['W-MAPE'] = world_mape

# Verify the result
print(low_technology_exports[['Country', 'Region', 'C-MAPE', 'R-MAPE', 'W-MAPE']].head())
```

```python
# Calculate the percentage of null values in C-APE for each country
c_gaps = low_technology_exports.groupby('Country')['C-APE'].apply(lambda x: x.isnull().mean() * 100).reset_index()

# Rename the new column to C-Gaps
c_gaps = c_gaps.rename(columns={'C-APE': 'C-Gaps'})

# Merge it with the original DataFrame
low_technology_exports = low_technology_exports.merge(c_gaps, on='Country', how='left')
```
